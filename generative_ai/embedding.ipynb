```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader("science_1007.pdf")
documents = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500,chunk_overlap = 100)
splitted_document = text_splitter.split_documents(documents)
splitted_document
```

### Ollama Embedding

```python
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="llama3.2",)
```

```python
embeddings
```

```python
query = "Dhoni is one of the best cricketer ever in cricket history"
em_qry = embeddings.embed_documents([query])
```

```python
len(em_qry[0])
```

### Huggingface Embedding

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['HUGGINGFACEHUB_API_TOKEN'] = os.getenv("HUGGINGFACEHUB_API_TOKEN")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-mpnet-base-v2")
```

```python
embeddings
```

```python
query = "Dhoni is one of the best cricketer ever in cricket history"
em_qry = embeddings.embed_query(query)
```

```python
len(em_qry)
```
