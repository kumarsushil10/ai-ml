# Pandas-DataFrame And Series
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
pd.__version__
```

```python
## Series
##A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series\n",series)
print(type(series))
```

```python
data = {'a':1,'b':2,'c':3}
print(pd.Series(data))
```

```python
data = [1,2,3,4,5]
index = ['a','b','c','d','e']
dataSeries = pd.Series(data=data,index=index)
print(dataSeries)
```

```python
# Dataframe from dictionary of list
data = {
    "name":["sushil","saket","manish","amri","anki"],
    "age":[29,35,34,19,23],
    "city":["Kolkata","Bangalore","pune","rachi","katra"]
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
# dataframe from list of dictionary

data = [
    {"name":"sushil","age":29,"city":"kolkata"},
    {"name":"saket","age":35,"city":"bangalore"},
    {"name":"manish","age":34,"city":"pune"},
]
df = pd.DataFrame(data)
print(df)
```
